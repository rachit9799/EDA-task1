```python
def payment(x):
    payData = x.split('₹')
    if (len(payData) !=1):
        return payData[1]
    else:
        return 0
    
def paymentMethod(x):
    payData = x.split('₹')
    return payData[0]
```

```python
lstCity = data['Shipping City'].unique()
```

```python
def cityFilter(x):
    city = x.split(' ')
    if(len(city) > 1):
        return city[0]
    return 0
```

```python
lstMainCity = []
for i in lstCity:
    ct = cityFilter(i)
    if  ct != 0:
        lstMainCity.append(ct)
```

```python
lstMainCity
```

```python
def clearSpaceProgramCity(x):
    city = x.split(' ')
    cityChange = False
    if len(city) > 1:
        for i in city:
            for j in lstMainCity:
                if i == j:
                    city = j
                    cityChange = True
                    break
    if cityChange ==True:
        return city
    else:
        strCity = ''
        for ele in city: 
            strCity += ele + ' '
        return strCity
```

```python
data['Shipping City'] = data['Shipping City'].apply(clearSpaceProgramCity)
data['Shipping City'] = data['Shipping City'].apply(lambda x: x.strip())
data['Shipping City'] =data['Shipping City'].str.replace("chennai-600113","chennai")
data['Shipping City'] =data['Shipping City'].str.replace("delhi.110092","delhi")
```
